```python
def get_annotations(filename, annotations_df):
    annots = annotations_df[annotations_df['Filename'] == filename]
    annot = []
    for i in range(len(annots)):
        x1 = annots.iloc[i]['Upper left corner X']
        y1 = annots.iloc[i]['Upper left corner Y']
        x2 = annots.iloc[i]['Lower right corner X']
        y2 = annots.iloc[i]['Lower right corner Y']
        annot.append([x1, y1, x2, y1, x2, y2, x1, y2])
    return annot
```

```python
def save_sample(sample_id, image, annotation):
    cv2.imwrite(f'./data/test_american/img_{sample_id}.jpg', image)
    with open(f'./data/test_american/img_{sample_id}.txt', 'w') as f:
        f.write(annotation)
```

```python
def process_sample(index, filename, annotations_df):
    img = cv2.imread('./data/data_unzipped/detection/american/' + filename)
    
    size = img.shape[:2]
    scale_ratio = 1024 / max(size)
    im = cv2.resize(img, (0, 0), fx=scale_ratio, fy=scale_ratio)
    annotations = []
    for annot in get_annotations(filename, annotations_df):
        annotations.append(','.join([str(int(a * scale_ratio)) for a in annot])+',')
    save_sample(index, im, '\n'.join(annotations))
```

```python
filenames = test_annotations['Filename'].unique()
for i, filename in tqdm.tqdm_notebook(enumerate(filenames), total=len(filenames)):
    process_sample(i, filename, test_annotations)
```
